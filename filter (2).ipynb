```python
import numpy as np
from scipy.signal import convolve2d
from skimage.data import coffee
from skimage import color
import matplotlib.pyplot as plt
```

```python
plt.imshow(coffee())
plt.show()
```

```python
img = color.rgb2gray(coffee())
```

```python
plt.imshow(img,cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
kenel_egde_detect1 = np.array([[1.,0.,-1.],
                                [0.,0.,0.],
                                [-1.,0.,1.]])
kenel_egde_detect2 = np.array([[0.,1.,0.],
                                [1.,-4.,1.],
                                [0.,1.,0.]])
kenel_egde_detect3 = np.array([[-1.,-1.,-1.],
                                [-1.,8.,-1.],
                                [-1.,-1.,-1.]])
kenel_egde_detect4 = np.array([[1.,0.,-1.],
                                [2.,0.,-2.],
                                [1.,0.,-1.]])
kenel_sharpen1 = np.array([[0.,-1.,0.],
                          [-1.,5.,-1.],
                          [0.,-1.,0.]])
kenel_sharpen2 = np.array([[-1.,-1.,-1.],
                           [-1.,9.,-1.],
                           [-1.,-1.,-1.]])
kenel_blur = np.array([[1.,1.,1.],
                       [1.,1.,1.],
                       [1.,1.,1.]])
kenel_egde_detect5 = np.array([[1.,2.,1.],
                                [0.,0.,0.],
                                [-1.,-2.,-1.]])
```

```python
def gray_show(img,kernel_spec):
  img_show = convolve2d(img,kernel_spec, "valid")
  plt.imshow(img_show, cmap= plt.cm.gray)
  plt.axis("off")
  plt.show()
```

```python
gray_show(img,kenel_egde_detect1)
```

```python
gray_show(img,kenel_egde_detect2)
```

```python
gray_show(img,kenel_egde_detect3)
```

```python
gray_show(img,kenel_egde_detect4)
```

```python
gray_show(img,kenel_egde_detect5)
```

```python
gray_show(img,kenel_sharpen1)
```

```python
gray_show(img,kenel_sharpen2)
```

```python
gray_show(img,kenel_blur)
```
